```python
import pandas as pd
import numpy as np

# Define price ranges per year
price_info = {2012:(1200,1500),2013:(1200,1500),2014:(1800,1800),2015:(1800,2000),
              2016:(2250,2250),2017:(2300,2300),2018:(2400,2400),2019:(2200,2500),
              2020:(2250,3000),2021:(2000,3250),2022:(3500,3500),2023:(3500,4000),2024:(3500,4500)}

export_info = {2012:(4000,6000),2013:(4200,6500),2014:(4500,6700),2015:(4700,6800),
               2016:(4800,7000),2017:(4900,7100),2018:(5000,7200),2019:(5100,7300),
               2020:(5200,7500),2021:(5300,7700),2022:(5400,7900),2023:(5500,8000),2024:(5600,8200)}

rainfall_info = {2012:(50,200),2013:(60,220),2014:(70,230),2015:(50,210),
                 2016:(60,250),2017:(70,260),2018:(60,240),2019:(50,220),
                 2020:(70,280),2021:(60,260),2022:(70,270),2023:(60,280),2024:(50,300)}

all_data = []
for year in price_info.keys():
    dates = pd.date_range(start=f'{year}-01-01', end=f'{year}-12-31', freq='W-MON')
    price_low, price_high = price_info[year]
    export_low, export_high = export_info[year]
    rain_low, rain_high = rainfall_info[year]

    for date in dates:
        price = price_low if price_low == price_high else np.random.randint(price_low, price_high+1)
        export_kg = np.random.randint(export_low, export_high+1)
        rainfall_mm = np.random.randint(rain_low, rain_high+1)
        all_data.append([date, price, export_kg, rainfall_mm])

df_synthetic = pd.DataFrame(all_data, columns=['Week', 'Price_Rs', 'Export_kg', 'Rainfall_mm'])
df_synthetic.to_csv("synthetic_cinnamon_data_full.csv", index=False)
print(df_synthetic.head())
```
